```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("data_csv.csv")
data.plot(kind="scatter",x="Year",y="Value")
plt.grid()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("data_csv.csv")
plt.figure(figsize=(10,3))
x=data["Value"]
plt.hist(x,bins=20,color="green")
plt.xlabel("Value")
plt.ylabel("Year")
plt.show
```
